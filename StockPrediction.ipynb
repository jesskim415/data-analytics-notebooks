```python
"""
Created on Tue Apr  2 17:03:53 2019

@author: jesskim
"""
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt #visualizing data
from sklearn.linear_model import LinearRegression # fitting model 
from sklearn.model_selection import train_test_split #train test split
from sklearn.metrics import r2_score # to sceck the score.

%matplotlib inline
```

```python
#set working directory
filename = '/Users/jesskim/Desktop/GOOGL.csv'
dataset = pd.read_csv(filename)
#changing the dataset into table. 
```

```python
#dependant and independant variables. 
prices = dataset['close'].tolist() # y - dependant value predict_value
initial = dataset['open'].tolist() # X - independant value (variables)
prices = np.reshape(prices, (len(prices), 1)) # making this into 1d vector
initial = np.reshape(initial, (len(initial), 1))
```

```python
#'Open'Graph

dataset[['open']].plot() 
plt.title('open Price')
plt.show()
```

```python
#'close' Graph

dataset[['close']].plot()
plt.title('Close Price')
plt.show()
```

```python
#training and test set. 
X_train, X_test, y_train, y_test = train_test_split(
        initial, prices, test_size = 0.2, random_state =0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#training set 
print('Train-set /','R2 score:',r2_score(y_train,regressor.predict(X_train)))
plt.scatter(X_train, y_train, color='red', label= 'Actual Price') 
plt.plot(X_train, regressor.predict(X_train), color='blue', linewidth=3, label = 'Predicted Price') 
plt.title('Linear Regression price| Open vs. Close')
plt.xlabel('Prices')
plt.legend()
plt.show()
```

```python
#Test set
print('Test-set/','R2 score:',r2_score(y_test,regressor.predict(X_test)))
plt.scatter(X_test, y_test, color='red', label= 'Actual Price')
plt.plot(X_test, regressor.predict(X_test), color='blue', linewidth=3, label = 'Predicted Price') 
plt.title('Linear Regression price| Open vs. Close')
plt.legend()
plt.xlabel('Prices')
plt.show()
```
